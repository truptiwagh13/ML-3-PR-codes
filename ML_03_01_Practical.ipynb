```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
k=10
num_problems=2000
q_star = np.random.normal(0,1,(num_problems,k))
arms=[0]*k
for i in range(10):
    arms[i]=np.random.normal(q_star[0,i],1,2000)
```

```python
plt.figure(figsize=(12,8))
plt.ylabel('Reward distribution')
plt.xlabel('Actions')
plt.xticks(range(1,11))
plt.yticks(np.arange(-5,5,0.5))
plt.violinplot(arms,positions=range(1,11),showmedians=True)
plt.show()
```

```python
import numpy as np
q_values = np.array([0.5,0.2])
action_greedy=np.argmax(q_values)
print(f"Q-values :  {q_values}")
print(f"Action selected by Greedy method : {action_greedy}")
```

```python
import random
q_values = np.array([0.5,0.2,0.3])
epsilon=0.3
if random.random()<epsilon:
  action_eg=random.choice([a for a in range(len(q_values)) if a != np.argmax(q_values)])
else:
  action_eg=np.argmax(q_values)

print(f"Q-values : {q_values}")
print(f"Epsilon : {epsilon}")
print(f"Action selected by Epsilon Greedy method: {action_eg}")
```
